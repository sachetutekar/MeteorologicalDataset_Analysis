##**Importing Necessary Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##**Importing Dataset**

```python
dataset = pd.read_csv('/content/WeatherHistory.csv')
dataset.head(6)
```

```python
dataset.describe() #describing the dataset
```

```python
#dropping the columns so that we get our required columns for further analysis and plotting of graphs
dataset = dataset.drop(['Summary', 'Precip Type', 'Temperature (C)', 'Wind Speed (km/h)', 
                        'Wind Bearing (degrees)', 'Visibility (km)', 'Loud Cover', 
                        'Pressure (millibars)', 'Daily Summary'], axis = 1) 
dataset.head(6)
```

```python
dataset.isnull().sum() #checking for summation of missing values
```

##**Converting Timezone to +00:00 UTC**

```python
dataset['Formatted Date'] = pd.to_datetime(dataset['Formatted Date'], utc = True)
dataset = dataset.set_index('Formatted Date')
req_data = dataset[['Apparent Temperature (C)', 'Humidity']].resample('MS').mean()
req_data.head(6)
```

```python
#checking for a specific month as an example
april = req_data[req_data.index.month == 4]
plt.figure(figsize = (15, 3))
plt.plot(april.loc['2006-04-01' : '2016-04-01', 'Apparent Temperature (C)'], marker = 'o', linestyle = 'dashed', label = 'Apparent Temperature (C)')
plt.plot(april.loc['2006-04-01' : '2016-04-01', 'Humidity'], marker = 'o', linestyle = 'dashed', label = 'Humidity')
plt.legend(loc = 'center right', fontsize = 15);
plt.xlabel('Month of April', fontsize = 15)
plt.title('Apparent Temperature (C) vs Humidity', fontsize = 15)
plt.xticks(fontsize = 15);
plt.yticks(fontsize = 13);
plt.show()
```

##**Plot the whole dataset for all months(together)**

```python
plt.figure(figsize=(15,3));
plt.plot(req_data['Humidity'], label = 'Humidity', color = 'orchid',linestyle='dashed');
plt.plot(req_data['Apparent Temperature (C)'], label = 'Apparent temp.',color = 'darkblue');
plt.title('Variation of Apparent temparature v/s Humidity', fontsize= 25);
plt.legend(loc = 0, fontsize = 12);
plt.xticks(fontsize = 15);
plt.yticks(fontsize = 13);
```
